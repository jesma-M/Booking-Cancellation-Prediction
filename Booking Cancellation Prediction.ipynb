```python
data.drop('Booking_ID',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['booking status']=data['booking status'].replace({'Canceled':0,'Not_Canceled':1})
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in ['type of meal','room type','market segment type']:
    data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
def convert_datetime(date_str):
    try:
        return pd.to_datetime(date_str, errors='raise', format='%m/%d/%Y')
    except ValueError:
        return pd.to_datetime(date_str, errors='coerce', format='%Y-%m-%d')
data['date of reservation'] = data['date of reservation'].apply(convert_datetime)
```

```python
data.head()
```

```python
data.info()
```

```python
data['reservation day']=data['date of reservation'].dt.day
data['reservation month']=data['date of reservation'].dt.month
data['reservation year']=data['date of reservation'].dt.year
```

```python
data.head()
```

```python
data.drop('date of reservation',axis=1,inplace=True)
```

```python
X=data.drop('booking status',axis=1)
y=data['booking status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
per=data['booking status'].value_counts(normalize=True)
```

```python
per
```

```python
def tuning(clf,parameter,X_train,y_train):
    cv=StratifiedKFold(n_splits=5,shuffle=True)
    cls_grid=GridSearchCV(clf,parameter,cv=cv,scoring='f1',n_jobs=-1)
    cls_grid.fit(X_train,y_train)
    best_hp=cls_grid.best_params_
    return cls_grid.best_estimator_,best_hp
```

```python
dt=DecisionTreeClassifier()
```

```python
dt_parameter_grid={'criterion':['gini','entropy'],
                  'max_depth':[12,13,14]}
```

```python
tuning(dt,dt_parameter_grid,X_train,y_train)
```

```python
dt=DecisionTreeClassifier(criterion='gini',max_depth= 12)
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
dtc_ac=accuracy_score(y_test,y_pred)
```

```python
dtc_ac
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
xgb=XGBClassifier()
```

```python
xgb_paran_grid={'n_estimators':[250,300],'learning_rate':[0.01,0.001,0.1],'max_depth':[7,8]}
```

```python
tuning(xgb,xgb_paran_grid,X_train,y_train)
```

```python
xgb=XGBClassifier(learning_rate= 0.1, max_depth=8, n_estimators= 300)
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred=xgb.predict(X_test)
```

```python
xgb_ac=accuracy_score(y_test,y_pred)
```
